```python
#importing necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
```

```python
# Loading Data
data = pd.read_excel(r"/content/drive/MyDrive/Colab Notebooks/DecisionTree_DataSet.xlsx")
data.head()
```

```python
#copying data
df = data.copy()
df.head()
```

```python
# For Information
df.info()
```

```python
# for describing
df.describe()
```

```python
# Setting Feature and Terget
X = df.drop(columns = ['species'],axis = 1)
y = df['species']
```

```python
#Converting categorical "species" as Numeric
y_encoded , y_label = pd.factorize(y)
print(y_encoded)
print(y_label)
```

```python
#Split daset

X_train, X_test , y_train , y_test = train_test_split( X , y_encoded , test_size = 0.25 , random_state = 42)
```

```python
# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Building and Trainning Model
model = DecisionTreeClassifier()

#traning The Model
model.fit(X_train_scaled , y_train)
```

```python
# Prediction
y_predict = model.predict(X_test_scaled)
print(y_predict)
```

```python
# Getting back the label
y_predict_label = y_label[y_predict]
print(y_predict_label)
```

```python
#Accuracy
accuracy = metrics.accuracy_score(y_test,y_predict)
confusion = metrics.confusion_matrix(y_test,y_predict)
classification_report = metrics.classification_report(y_test , y_predict)



print(f'Accuracy : {accuracy}')
print(f'Confusion : {confusion}')
print(f'Classification Report : {classification_report}')
```

```python
# Prediction
y_predict = model.predict(X_test_scaled)
print(y_predict)
```
